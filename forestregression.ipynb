```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
```

```python
## df_copy Monthly fire analysis for Bejia
dftemp = df[df['Region']==0]
sns.set_style('whitegrid')
sns.countplot(dftemp,x = 'month', hue = 'Classes')
plt.title('Fire analysis Bejia')
plt.xlabel('Month')
plt.ylabel('No. of fires')
plt.legend()
plt.show()
```

```python
dftemp1 = df[df['Region']==1]
sns.set_style('whitegrid')
sns.countplot(dftemp1,x = 'month', hue = 'Classes')
plt.title('Fire analysis Sidi-Bel')
plt.xlabel('Month')
plt.ylabel('No. of fires')
plt.legend()
plt.show()
```
